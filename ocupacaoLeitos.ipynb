```python
# DATASET ORIGINAL ############################################################################################################
```

```python
# Coleta
dados = pd.read_csv("ocupacaoLeito.csv",parse_dates = ['dataNotificacao'])

# Primeiros 5 regitros
dados.head()
```

```python
# DATASET MANIPULADO ###########################################################################################################
```

```python
#Expurgados os registros que foram excluídos, filtrando apenas os que apresentam FALSO na coluna 'excluido'
dadosf = dados.loc[dados['excluido'] == 0]

dadosf.head()
```

```python
# Informações sobre o DataFrame data_f
dadosf.info()
```

```python
# TRATAMENTO ###################################################################################################################
```

```python
# Filtradas as colunas necessárias
dadosf = dadosf[["dataNotificacao", "ocupacaoSuspeitoUti", "ocupacaoConfirmadoUti","saidaSuspeitaObitos", "saidaConfirmadaObitos", "estadoNotificacao", "municipioNotificacao"]]
```

```python
dadosf.info()
```

```python
dadosf
```

```python
# Verificação dos itens nulos
dadosf.isnull().sum()
```

```python
# Substituição dos registros nulos por 0
dadosf[['ocupacaoSuspeitoUti','ocupacaoConfirmadoUti','saidaSuspeitaObitos', 'saidaConfirmadaObitos']] = dadosf[['ocupacaoSuspeitoUti','ocupacaoConfirmadoUti','saidaSuspeitaObitos', 'saidaConfirmadaObitos']].fillna(0)

# Exclusão das linhas com registros nulos (em função das datas de notificação não preenchidas)
dadosf.dropna(inplace=True)

dadosf.isnull().any()
```

```python
# Tratamento da data e hora de notificação, restringindo à data
dadosf.dataNotificacao = dadosf.dataNotificacao.dt.date
dadosf.dataNotificacao = pd.to_datetime(dadosf.dataNotificacao)
```

```python
dadosf.dataNotificacao.describe()
```

```python
# Definição dos tipos das variáveis
dadosf.ocupacaoSuspeitoUti = dadosf.ocupacaoSuspeitoUti.astype(int)
dadosf.ocupacaoConfirmadoUti = dadosf.ocupacaoConfirmadoUti.astype(int)
dadosf.saidaSuspeitaObitos = dadosf.saidaSuspeitaObitos.astype(int)
dadosf.saidaConfirmadaObitos = dadosf.saidaConfirmadaObitos.astype(int)
```

```python
dadosf.info()
```

```python
# Exibição da descrição do campo 
dadosf.dataNotificacao.describe()
```

```python
# Ordenação do dataframe pela data, de forma decrescente
dadosf = dadosf.sort_values(by='dataNotificacao', ascending=False)
dadosf
```

```python
# Filtrando registros que possuam data igual ou maior que 01/01/2020
dadosf.query('dataNotificacao >= "2020-01-01"', inplace=True)
```

```python
dadosf
```

```python
dadosf.dataNotificacao.describe()
```

```python
# Tratamento do valor duplicado encontrado
dadosf = dadosf.replace(['GOIAS'], 'Goiás')
```
